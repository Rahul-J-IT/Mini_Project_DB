```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import current_timestamp, lit, input_file_name
 
spark = SparkSession.builder.getOrCreate()


spark.sql("CREATE SCHEMA IF NOT EXISTS novocart.bronze")
 
tables = ["customers", "products", "orders", "order_items", "exchange_rates"]
 
for tbl in tables:
    df = spark.table(f"azure_blob_storage.{tbl}")
    df.write.mode("overwrite").saveAsTable(f"novocart.bronze.{tbl}_raw")
    print(f"Bronze: {tbl}_raw loaded with {df.count()} rows")
```
